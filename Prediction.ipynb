```python
count_null_columns = []
count_null = []
for i in df.columns:
    if sum(df[i].isnull() == True) > 0:
        count_null_columns.append(i)
        count_null.append(sum(df[i].isnull() == True))
```

```python
count_null = {}
for i in df.columns:
    if sum(df[i].isnull() == True) > 0:
        count_null[i] = sum(df[i].isnull() == True)
```

```python
count_null_sorted = dict(sorted(count_null.items(), key=lambda item: item[1]))
```

```python
count_null_df = pd.DataFrame ({
        'Group':  count_null_sorted.keys(),
        'Value': count_null_sorted.values()
})
```

```python
plt.figure(figsize=(10,7))
plt.barh(y=count_null_df['Group'], width=count_null_df['Value'])
plt.title('Null Values Count')
plt.grid()
plt.savefig('Null Values Count.png')
```

```python
# Create an object_list to store columns that has object dtype
object_list = []
for i in df.columns:
    if df[i].dtype == object:
        object_list.append(i)
```

```python
# Create a dummies dictionary to contain sub-dataframes which has been converted to dummies
dummies_dict = {}
for i in object_list:
    dummies_dict[i] = pd.get_dummies(df[i],drop_first=True)
```

```python
# Concatenate all sub-dataframes in dummies
dummies = pd.concat([dummies_dict[i] for i in object_list], axis=1)
```

```python
dummies.head()
```

```python
# Drop columns that have had converted by dummies
for i in object_list:
    df.drop(i,axis=1,inplace=True)
```

```python
df.info()
```

```python
# Concatenate the previous processed dataframe with the dummies dataframe
df_with_dummies = pd.concat([df,dummies],axis=1)
```

```python
df_with_dummies.info()
```

```python
df_with_dummies.head()
```

```python
df_with_dummies.columns
```

```python
# Fill null values to apply Machine Learning
for i in df_with_dummies.columns:
    df_with_dummies[i].fillna(value=round(df_with_dummies[i].mean()))
```

## Apply Machine Learning for Clean Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_with_dummies.drop(['Id','SalePrice'],axis=1)[:1460], df_with_dummies['SalePrice'][:1460], test_size=0.4, random_state=42)
```

### Choose LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
prediction = lm.predict(X_test)
```

```python
sns.jointplot(x=y_test,y=prediction,ylim=(0,600000),xlim=(0,600000),kind="reg")
plt.suptitle('y_test vs. prediction', y = 1)
plt.savefig('y_test vs. prediction.png')
```

```python
sns.histplot((y_test-prediction),bins=200)
plt.xlim(-100000,200000)
plt.title('Error Distribution')
plt.savefig('Error Distribution.png')
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```
